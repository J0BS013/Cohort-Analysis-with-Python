```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import warnings
warnings.filterwarnings("ignore")
```

#### Loading Transaction Data into a Dataframe

```python
df = pd.read_excel("Online Retail.xlsx")
```

```python
df.head()
```

#### Data Wrangling

```python
#Cleaning Data
#Droping rows with no CustomerID

df = df.dropna(subset=['CustomerID'])
df.info()
```

```python
#Getting only the month and year of the Invoice Date

#Function for month
def get_month(x):
    return datetime.datetime(x.year, x.month,1)

df['InvoiceMonth'] = df['InvoiceDate'].apply(get_month)

df.head()
```

#### Separating months into Cohort indexes

```python
#Cohort Month is the first time customer was acquired (Customer Birth)

df['Cohort Month'] =  df.groupby('CustomerID')['InvoiceMonth'].transform('min')

df.tail(10)
```

```python
#Create a date element function to get a series for cohort calculation

def get_date_elements(df, column):
    day = df[column].dt.day
    month = df[column].dt.month
    year = df[column].dt.year
    return day, month, year 
```

```python
#Get date elements for our cohort and invoice columns

Invoice_day,Invoice_month,Invoice_year =  get_date_elements(df,'InvoiceMonth')
Cohort_day,Cohort_month,Cohort_year =  get_date_elements(df,'Cohort Month')
```

```python
#Create a Cohort Index
#Cohort index is the month difference between invoice month and cohort month for each row.

year_diff = Invoice_year - Cohort_year
month_diff = Invoice_month - Cohort_month
df['CohortIndex'] = year_diff*12+month_diff +1

df.tail()
```

```python
#Count the customer ID by grouping by Cohort Month  and Cohort Index

cohort_data = df.groupby(['Cohort Month','CohortIndex'])['CustomerID'].apply(pd.Series.nunique).reset_index()
cohort_data
```

```python
#Pivot Table for Cohort Table

cohort_table = cohort_data.pivot(index='Cohort Month', columns=['CohortIndex'],values='CustomerID')
cohort_table
```

#### Cohort Retention Charts

```python
#Changing Date Format

cohort_table.index = cohort_table.index.strftime('%B %Y')
```

```python
plt.figure(figsize=(23,12))
sns.heatmap(cohort_table,annot=True,cmap='Blues')
plt.show()
```

```python
# Cohort Percentage Table

Percent_Cohort_Table = cohort_table.divide(cohort_table.iloc[:,0],axis=0)
Percent_Cohort_Table
```

```python
plt.figure(figsize=(23,12))
sns.heatmap(Percent_Cohort_Table,annot=True,cmap='Blues', fmt='.0%')
plt.show()
```
